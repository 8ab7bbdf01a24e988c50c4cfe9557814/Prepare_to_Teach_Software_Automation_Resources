# Data Graphs used in presentation

#### Import Dependancies

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Linear Relationship w Adjustable Correlation
correlation = 0.8       # Correlation between -1 and 1
n = 300                 # Sample

xx = np.array([0, 50])
yy = np.array([0, 50])
means = [xx.mean(), yy.mean()]  
stds = [xx.std() / 3, yy.std() / 3]

covs = [[stds[0]**2          , stds[0]*stds[1]*correlation], 
        [stds[0]*stds[1]*correlation,           stds[1]**2]] 

m = np.random.multivariate_normal(means, covs, n).T
plt.xlabel("Feature")
plt.ylabel("Target")
plt.scatter(m[0], m[1], color='blue')
plt.show()
```

```python
# Polynomial Linear Relationship
n = 300 # Sample
Beta = [0, 1, -2, 0.5]
e = np.random.normal(-3, 4, n)
X = 2 - 3 * np.random.normal(0, 0.5, n)
y = 0
for i in range(len(Beta)):
    y += Beta[i] * X**i
y += e

plt.xlabel("Feature")
plt.ylabel("Target")
plt.scatter(X, y, s=10)
plt.show()
```

```python
x = np.linspace(-10, 10, 800)
y = x**2 + 0.1*x + 2  

plt.xlabel("Parameters (Coef & y_intercept)")
plt.ylabel("Cost")
plt.plot(x, y)
plt.show()
```

y = mx + b, where 'm' is the slope and 'b' is the y-intercept.
