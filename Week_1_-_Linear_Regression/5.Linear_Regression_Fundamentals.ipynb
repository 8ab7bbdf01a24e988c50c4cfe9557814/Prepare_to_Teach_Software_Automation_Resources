A simple linear function

```math
y = mx + b
```

| Symbol | |
| --- | --- |
| y | Predicted value |
| m | Slope of coefecient of the linear expression |
| x | Feature |
| b | y-intercept when the line corsses the y axis |


````python
x = np.array([[2], [4], [6], [8], [10], [12], [14], [16]])
y = np.array([1, 3, 5, 7, 9, 11, 13, 15])
````

````python
def compute_model_output(x, m, b):
    X = x.shape[0]
    y_pred = np.zeros(X)
    for i in range(X):
        y_pred[i] = m * x[i].item() + b
    return y_pred
````

````python
m = 0
b = 0

tmp_f_wb = compute_model_output(x, m, b,)

# Plot our model prediction
plt.plot(x, tmp_f_wb, c='b',label='Prediction')

# Plot the data points
plt.scatter(x, y, marker='x', c='r',label='Actual Values')

# Set the title
plt.title("Course Specifications")
# Set the y-axis label
plt.ylabel('Features')
# Set the x-axis label
plt.xlabel('Target')
plt.legend()
plt.show()
````
